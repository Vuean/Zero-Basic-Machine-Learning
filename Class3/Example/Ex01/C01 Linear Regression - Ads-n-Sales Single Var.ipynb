### 3.2.5 数据清洗与规范化

```python
# 构建特征集，只含有微信广告一个特征
X = np.array(df_ads.wechat)
# 构建标签集
y = np.array(df_ads.sales)

print("张量X的阶：", X.ndim)
print("张量X的形状：", X.shape)
print("张量X的内容：", X)
```

```python
# 通过reshape函数把向量转换为矩阵，len函数返回样本个数
X = X.reshape(len(X), 1)
y = y.reshape(len(y), 1)

print("张量X的形状：", X.shape)
print("张量X的内容：", X)
```

### 3.2.6 拆分数据集为训练集和测试集

```python
# 将数据集进行80%训练集和20%测试集的分割：
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                                                                    random_state=0)
```

### 3.2.7 数据归一化

```python
def scaler(train, test):
    min = train.min(axis=0)
    max = train.max(axis=0)
    gap = max - min
    train -= min
    train /= gap
    test -= min
    test /= gap
    return train, test
```

```python
X_train, X_test = scaler(X_train, X_test)
y_train, y_test = scaler(y_train, y_test)
```

```python
plt.plot(X_train, y_train, 'r.', label='Training data')
plt.xlabel('Wechat Ads')
plt.ylabel('Sales')
plt.legend()    # 显示图例
plt.show()
```

### 3.3.3 算是误差函数——Lwb

```python
### 3.3.3 算是误差函数——Lwb
def loss_function(X, y, weight, bias):
    y_hat = weight * X + bias
    loss = y_hat - y
    cost = np.sum(loss**2)
    return cost
```

```python
print ("当权重5，偏置3时，损失为：", 
loss_function(X_train, y_train, weight=5, bias=3))
print ("当权重100，偏置1时，损失为：", 
loss_function(X_train, y_train, weight=100, bias=1))
```

### 3.4.4 学习速率也很重要

```python
def gradient_descent(X, y, w, b, lr, iter):
    l_history = np.zeros(iter)
    w_history = np.zeros(iter)
    b_history = np.zeros(iter)
    for i in range(iter):
        y_hat = w*X + b
        loss = y_hat - y
        derivative_w = X.T.dot(loss) / len(X)
        derivative_b = sum(loss)*1 / len(X)
        w = w - lr * derivative_w
        b = b - lr * derivative_b
        l_history[i] = loss_function(X, y, w, b)
        w_history[i] = w
        b_history[i] = b
    return l_history, w_history, b_history
```

## 3.5 实现一元线性回归模型并调试超参数

### 3.5.1 权重和偏置的初始值

```python
# 首先确定参数的初始值
iterations = 225;   # 迭代250次
alpha = 0.5;
weight = -4
bias = 3
print('当前损失：', loss_function(X, y, weight, bias))
```

```python
# 绘制当前的函数模型
plt.plot(X_train, y_train, 'r.', label='Training data')
line_X = np.linspace(X_train.min(), X_train.max(), 500)
line_y = [weight*xx + bias for xx in line_X]
plt.plot(line_X, line_y, 'b--', label='Current hypothesis')
plt.xlabel('Wechat Ads')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

### 3.5.2 进行梯度下降

```python
# 根据初始参数值，进行梯度下降，也就是开始训练机器，拟合函数
loss_history, weight_history, b_history = gradient_descent(
    X_train, y_train, weight, bias, alpha, iterations)
```

```python
plt.plot(loss_history, 'g--', label='Loss Curve')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

### 3.5.3 调试学习速率

```python
# 绘制当前的函数模型
plt.plot(X_train, y_train, 'r.', label='Training Data')
line_X = np.linspace(X_train.min(), X_train.max(), 500)
lint_y = [weight_history[-1]*xx + b_history[-1] for xx in line_X]
plt.plot(line_X, lint_y, 'b--', label='Current hypothesis')
plt.xlabel('Wechat Ads')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
print('当前损失：', loss_function(X_train, y_train,
                                                                weight_history[-1], b_history[-1]))
print('当前权重：', weight_history[-1])
print('当前偏置：', b_history[-1])
```

```python
print('测试集损失：', loss_function(X_test, y_test, 
                                        weight_history[-1], b_history[-1]))
```
